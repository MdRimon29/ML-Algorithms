```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
digit=load_digits()
```

```python
dir(digit)
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(digit.data,digit.target,train_size=0.8,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

Predict for first 6 data from digit dataset

```python
model.predict(digit.data[:6])
```

Create a confusion matrix so that we can find where is the wrong prediction

```python
y_predicted=model.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

Visualizing it by using seaborn 

```python
import seaborn as sn
plt.figure(figsize = (5,3))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
