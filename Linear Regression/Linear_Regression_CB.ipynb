```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('linear_regression_data.csv')
plt.xlabel("Weight (gm)")
plt.ylabel("Price ($)")
```

```python
plt.scatter(df['weight'],df['price'])
plt.show()
```

```python
model=LinearRegression()
model.fit(df[['weight']],df.price)
```

```python
print(model.predict([[100]]))
print(model.coef_)
print(model.intercept_)
print(model.coef_*100+model.intercept_)
```

```python
d=pd.read_csv('linear_regression_data2.csv')
print(d.head(3))
p=model.predict(d)
print (p)
```

```python
d['price']=p    #all predicted values added to our csv file
d.to_csv("predictions_linear_regression.csv") #to export to a new csv file called predictions.csv
```
