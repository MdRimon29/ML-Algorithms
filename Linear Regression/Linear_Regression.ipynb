Import section

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Dataset section

```python
df=pd.read_csv('linear_regression_data.csv')

plt.xlabel("weight(gm)")
plt.ylabel('price')

X=df[['weight']].values
Y=df['price'].values
```

Linear Regression

```python
model=LinearRegression()
model.fit(X,Y)  #if we  define x as x=df['weight'].values then we need to write here [[i] for i in x] instead of x
coef=model.coef_
intrecept=model.intercept_
```

Prediction

```python
x_new=100
y_new=coef*x_new+intrecept

print(y_new)
```

```python
plt.scatter(X,Y)
plt.plot(X, model.predict(X), color='red')
plt.show()
```
